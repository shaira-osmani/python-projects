# Bank Management System GUI app - Python

```python
# Import the libreries
import tkinter as tk
from tkinter import messagebox
from time import gmtime, strftime

# Check if the values are typed correctly
def is_number(s):
    try:
        float(s)
        return 1
    except ValueError:
        return 0

# Check if the account is registered.
def check_acc_numb(num):
    try:
        fpin=open(num+".txt","r")
    except FileNotFoundError:
        messagebox.showinfo("Error", "Invalid Credintials!\n Try again")
        return 0
    fpin.close()
    return

def home_return(master):
    master.destroy()
    Main_Menu()

def write(master,name,oc,pin):
    if((is_number(name)) or (is_number(oc) == 0) or (is_number(pin)==0) or name ==""):
        messagebox.showinfo("Error","Invalid Credentials\n Please try again!")
        master.destroy()
        return

    user_file = open("Account_Record.txt","r")
    account_no = int(user_file.readline())
    account_no +=1
    user_file.close()

    user_file = open("Account_Record.txt","w")
    user_file.write(str(account_no)+".txt","w")
    user_file.close()

    user_details = open(str(account_no)+".txt","w")
    user_details.write(pin+"\n")
    user_details.write(oc+"\n")
    user_details.write(str(account_no)+"\n")
    user_details.write(name+"\n")
    user_details.close()

    user_transaction = open(str(account_no)+"-rec.txt","w")
    user_transaction.write("Date Credit Debit Balance\n")
    user_transaction.write(str(strftime("[%Y-%m-%d][%H:%M:%S]", gmtime()))+ " " + oc + " " + oc +"\n")
    user_transaction.close()

    messagebox.showinfo("Details","Your account number is: "+str(account_no))
    master.destroy()
    return

def credit_write(master,user_input,account,name):
    if(is_number(user_input)==0):
        messagebox.showinfo("Error","Invalid Credentials")
        master.destroy()
        return
    
    user_details = open(account+".txt","r")
    file_output = user_details.readline()
    account_balance = int(user_details.readline())
    user_details.close()

    user_input = int(user_input)
    credit_balance = user_input + account_balance
    user_details = open(account+".txt","w")
    user_details.write(file_output)
    user_details.write(credit_balance+"\n")
    user_details.write(account+"\n")
    user_details.write(name+"\n")
    user_details.close()

    user_transaction = open(str(account)+"-rec.txt","a+")
    user_transaction.write(str(strftime("[%Y-%m-%d][%H:%M:%S]", gmtime()))+ " " + str(user_input) + " " + str(credit_balance) +"\n")
    user_transaction.close()

    messagebox.showinfo("Operation Successfull!" , "Amount Credited Successfully!")
    master.destroy()
    return
    
def debit_write(master,user_input,account,name):
    if(is_number(user_input)==0):
        messagebox.showinfo("Error","Invalid Credintials")
        master.destroy()
        return
    user_details = open(account+".txt","r")
    file_output = user_details.readline()
    account_balance = int(user_details.readline())
    user_details.close()

    if(int(user_input) > account_balance):
        messagebox.showinfo("Error!","You dont have enough amount left in your account.")
    else:
        user_data = int(user_input)
        credit_balance = account_balance - user_data
        user_details = open(account+".txt","w")
        user_details.write(file_output)
        user_details.write(str(credit_balance)+"\n")
        user_details.write(account+"\n")
        user_details.write(name+"\n")
        user_details.close()

        user_transaction = open(str(account)+"-rec.txt","a+")
        user_transaction.write(str(strftime("[%Y-%m-%d][%H:%M:%S]", gmtime()))+ " " + str(user_input) + " " + str(credit_balance) +"\n")
        user_transaction.close()

        messagebox.showinfo("Operation Successfull!" , "Amount Credited Successfully!")
        master.destroy()
        return

# Login Menu

def logged_in_menu(accnt,name):
	rootwn=tk.Tk()
	rootwn.geometry("1600x500")
	rootwn.title("CopyAssignment Bank | Welcome - "+name)
	rootwn.configure(background='SteelBlue1')
	fr1=tk.Frame(rootwn)
	fr1.pack(side="top")
	l_title=tk.Message(rootwn,text="BANK MANAGEMENT SYSTEM",relief="raised",width=2000,padx=600,pady=0,fg="white",bg="blue4",justify="center",anchor="center")
	l_title.config(font=("Arial","50","bold"))
	l_title.pack(side="top")
	label=tk.Label(text="Logged in as: "+name,relief="raised",bg="blue3",font=("Times",16),fg="white",anchor="center",justify="center")
	label.pack(side="top")
	img2=tk.PhotoImage(file="credit.gif")
	myimg2=img2.subsample(2,2)
	img3=tk.PhotoImage(file="debit.gif")
	myimg3=img3.subsample(2,2)
	img4=tk.PhotoImage(file="balance1.gif")
	myimg4=img4.subsample(2,2)
	img5=tk.PhotoImage(file="transaction.gif")
	myimg5=img5.subsample(2,2)
	b2=tk.Button(image=myimg2,command=lambda: Cr_Amt(accnt,name))
	b2.image=myimg2
	b3=tk.Button(image=myimg3,command=lambda: De_Amt(accnt,name))
	b3.image=myimg3
	# b4=tk.Button(image=myimg4,command=lambda: disp_bal(accnt))
	# b4.image=myimg4
	# b5=tk.Button(image=myimg5,command=lambda: disp_tr_hist(accnt))
	# b5.image=myimg5
	
	img6=tk.PhotoImage(file="logout.gif")
	myimg6=img6.subsample(2,2)
	b6=tk.Button(image=myimg6,relief="raised",command=lambda: logout(rootwn))
	b6.image=myimg6
	b2.place(x=100,y=150)
	b3.place(x=100,y=220)
	b4.place(x=900,y=150)
	b5.place(x=900,y=220)
	b6.place(x=500,y=400)
def check_log_in(master,name,acc_num,pin):
	if(check_acc_nmb(acc_num)==0):
		master.destroy()
		Main_Menu()
		return

	if( (is_number(name))  or (is_number(pin)==0) ):
		messagebox.showinfo("Error","Invalid Credentials\nPlease try again.")
		master.destroy()
		Main_Menu()
	else:
		master.destroy()
		logged_in_menu(acc_num,name)

# Login into system
def login(master):
    master.destroy()
    login_window = tk.Tk()
    login_window.geometry("600x500")
    login_window.title("Login into system")
    login_window.configure(bg="white")
    frame1 = tk.Frame(login_window,bg="blue")
    frame1.pack(side="top")
    i_title = tk.Message(login_window,text="BANK MANAGEMENT SYSTEM", relief="raised", width=2000,padx=600,pady=0,fg="white",bg="blue4",justify="center",anchor="center")
    i_title.config(font=("Ariel","50","bold"))
    i_title.pack(side="top")
    label1 = tk.Label(text="Enter Name", relief="raised",font=("Times",16))
    label1.pack(side="top")
    entry1 = tk.Entry(login_window)
    entry1.pack(side="top")
    label2 = tk.Label(text="Enter Account Number", relief="raised",font=("Times",16))
    label2.pack(side="top")
    entry2 = tk.Entry(login_window)
    entry2.pack(side="top")
    label3 = tk.Label(text="Enter Pin Number", relief="raised",font=("Times",16))
    label3.pack(side="top")
    entry3 = tk.Entry(login_window,show="")
    entry3.pack(side="top")

    submit_btn = tk.Button(login_window, text="Submit", command=lambda:check_log_in(login_window,entry1.get().strip(),
                           entry2.get().strip(), entry3.get().strip()))
    submit_btn.pack(side="top")

    home_btn = tk.Button(text="Home",font=("Times",16), relief="raised", bg="blue4", fg="white",command=lambda: home_return(login_window))
    home_btn.pack(side="top")
    login.bind("<Return>", lambda x:check_log_in(login_window, entry1.get().strip(),entry2.get().strip(), entry3.get().strip()))

def logout(master):
	
	messagebox.showinfo("Logged Out","You Have Been Successfully Logged Out!!")
	master.destroy()
	Main_Menu()

def check_log_in(master,name,acc_num,pin):
	if(check_acc_nmb(acc_num)==0):
		master.destroy()
		Main_Menu()
		return

	if( (is_number(name))  or (is_number(pin)==0) ):
		messagebox.showinfo("Error","Invalid Credentials\nPlease try again.")
		master.destroy()
		Main_Menu()
	else:
		master.destroy()
		logged_in_menu(acc_num,name)
```
