# YouTube download Manager with Python - GUI

```python
#import the libraries
from tkinter import*
from pytube import YouTube
```

```python
#initialize tkinter
root = Tk()
#set the gui diemensions
root.geometry("400x350")
root.resizable(0, 0) # makes the window adjustable with its features

#set the gui title
root.title("Youtube Download Manager")


#download function
def download():
    #use try and catch to run the program without breaking down
    try:
        myVar.set("Downloading...")
        root.update()
        url = YouTube(str(link.get()))
        video = url.streams.first()
        video.download()
        link.set("Video downloaded successfully")

    except Exception as e:
        myVar.set("Mistake")
        root.update()
        link.set("Enter correct link")
    
#create the lable widget to welcome user
Label(root, text="Welcome to Youtube download manager", font="Consolas 15 bold").pack()
myVar = StringVar()
myVar.set("Enter the link below")
Entry(root, textvariable=myVar, width=40).pack(pady=10)
link = StringVar()
Entry(root, textvariable=link, width=40).pack(pady=10)
Button(root, text="Download Video", command=download).pack()
root.mainloop()
```
